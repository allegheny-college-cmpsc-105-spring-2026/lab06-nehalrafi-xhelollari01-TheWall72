```python
# Survey

# : Do one or more explorations with dfs
# no interaction is required

# make a blank canvas

# make a title starting with 'Figure 2:' Fill in a relevant title

counts = dfs["Which House Is Better"].value_counts()

# Plot
fig, ax = plt.subplots(figsize=(7, 5))

colors = ["#C8102E", "#003087", "#F5A623"]
bars = ax.bar(counts.index, counts.values, color=colors, edgecolor="white", linewidth=0.8)
for bar in bars:
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.1,
        str(int(bar.get_height())),
        ha="center", va="bottom", fontsize=12, fontweight="bold"
    )

ax.set_title("Figure 2: Which House Is Better?", fontsize=16, fontweight="bold", pad=15)
ax.set_xlabel("House", fontsize=12)
ax.set_ylabel("Number of Votes", fontsize=12)
ax.set_yticks(range(0, counts.max() + 2))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()

# etc.

# reveal everything
```
